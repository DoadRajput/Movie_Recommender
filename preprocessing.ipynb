# Importing libraries for preprocessing

```python
import pandas as pd
import ast
```

# Loading CSV files

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
credits.head()
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv") 
movies.head()
```

# Preprocessing

```python
df = movies.merge(credits,on='title')
df.head(1)
```

```python
df = df[['title','overview','genres','keywords','cast','crew','movie_id']]
df.head(1)
```

```python
df.dropna(inplace=True)
```

```python
def convert(text):
    return [item['name'] for item in ast.literal_eval(text)]
```

```python
df['genres'] = df['genres'].apply(convert)
df['keywords'] = df['keywords'].apply(convert)
df['cast'] = df['cast'].apply(convert)
```

```python
df['cast'] = df['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
df.head(1)
```

```python
def collapse(L):
    L1 = []
   
    for i in L:
        L1.append(i.replace(" ",""))
    L1 = ', '.join(L1)
    return L1
```

```python
df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
```

```python
df.head(1)
```

```python
df['feature'] = df['overview'] +' '+ df['genres'] + ' '+ df['keywords'] + ' '+ df['cast'] +' '+ df['crew']
```

```python
df['feature'] = df['feature'].str.lower()
df['title'] = df['title'].str.lower()
```

```python
essential_col = ['title','feature','movie_id']
df = df[essential_col]
```

# Uploading preprocessed CSV file

```python
df.to_csv("movies.csv", index=False)
```
